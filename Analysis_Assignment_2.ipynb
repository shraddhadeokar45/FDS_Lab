```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = data = {
    'Product': ['Laptop', 'Smartwatch', 'Laptop', 'Smartwatch', 'Camera', 'Headphones', 'Laptop', 'Camera'],
    'Category': ['Electronics', 'Wearable', 'Electronics', 'Wearable', 'Photography', 'Audio', 'Electronics', 'Photography'],  # Qualitative
    'Brand': ['Dell', 'Samsung', 'Dell', 'Apple', 'Canon', 'Sony', 'HP', 'Canon'],  # Qualitative
    'Units_Sold': [120, 340, 150, 90, 75, 210, 65, 130],  # Quantitative
    'Unit_Price': [55000, 30000, 55000, 50000, 20000, 8000, 12000, 20000]  # Quantitative
}

df = pd.DataFrame(data)
df.head()
```

```python
# --- QUALITATIVE ANALYSIS ---
print("Qualitative Analysis:")
print(df['Product'].value_counts())      # Frequency of categories
print(df['Brand'].value_counts())  # Frequency of departments

# Bar chart for Gender distribution
df['Product'].value_counts().plot(kind='bar', color=['lightblue', 'lightgreen'])
plt.title("Product Sell Analysis")
plt.xlabel("Product")
plt.ylabel("Brand")
plt.show()
```

```python
# --- QUANTITATIVE ANALYSIS ---
print("\nQuantitative Analysis:")
print(df[['Units_Sold', 'Unit_Price']].describe())  # Summary statistics

# Histogram for Age
df['Units_Sold'].plot(kind='hist', bins=5, color='lightblue', edgecolor='black')
plt.title("Sell Analysis")
plt.xlabel("Sold")
plt.ylabel("Cost")
plt.show()

# Scatter plot: Age vs Salary
plt.scatter(df['Units_Sold'], df['Unit_Price'], color='green')
plt.title("Sell Analysis")
plt.xlabel("Sold")
plt.ylabel("Cost")
plt.show()
```
